# **Session 2: Text Visualization**

Text Visualization is the process of representing textual data graphically, allowing patterns, trends, and insights to be more easily recognized and understood. Instead of simply reading through text data, text visualization enables you to see the relationships, frequency of terms, or the structure of the data in a visual format.

### **Common Text Visualization Techniques:**

1. **Word Clouds:**
   - A simple yet effective visualization that highlights the most frequent words in a dataset. The size of each word corresponds to its frequency or significance in the text.

2. **Bar Charts:**
   - These charts illustrate the frequency or distribution of words, topics, or sentiments within the text. For instance, a bar chart can be used to show the top 10 most common words in a document.

3. **Histograms:**
   - Similar to bar charts, histograms are often used to display distributions, such as the length of words or sentences within a text corpus.

```python
corpus = open('/content/sample_data/ObamaSpeech.txt', 'r').read()
print(corpus)
```

```python
import matplotlib.pyplot as ptl
import wordcloud
import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

## **Task 1: Pre-processing the text for visualisation**

To prepare your corpus for effective text visualization, follow these steps:

1. **Convert to Lowercase:**
   - Transform all characters in your text to lowercase using the `lower()` function. This standardizes the text and ensures that words are treated consistently, regardless of their case. For example: `corpus.lower()`.

2. **Word Tokenization:**
   - Break down the text into individual words by applying word tokenization. This process splits the text into a list of words, making it easier to analyze and visualize word frequencies and patterns. For example, you can use `nltk.word_tokenize(text)` to achieve this.

```python
words=nltk.word_tokenize(corpus.lower())

print(words)
```

3. **Remove Punctuation:**
   - Eliminate punctuation from each word by filtering out any non-alphabetic characters. This can be done by ensuring that each word contains only alphabetic characters. For example, use `words = [word.lower() for word in words if word.isalpha()]` to create a list of words with punctuation removed.

```python
# remove punctuation from each word
words=[word.lower() for word in words if word.isalpha()]
words
```

4. **Remove Stop Words:**

   Eliminate common stop words from your list to concentrate on more meaningful terms. Stop words are frequently used words like "and," "the," or "is," which often do not contribute substantial meaning to the analysis. By removing these, you can better focus on the words that provide valuable insights.

```python
from nltk.corpus import stopwords

filtered_sentence=[]

stop_words=set(stopwords.words("english"))
for w in words:
    if w not in stop_words:
        filtered_sentence.append(w)

print(filtered_sentence)
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
nltk.download('omw-1.4')
nltk.download('wordnet')

lemmatizer=WordNetLemmatizer()

lemmatized_words = []

for w in filtered_sentence:
    rootWord = lemmatizer.lemmatize(w,pos=wordnet.VERB)
    lemmatized_words.append(rootWord)

print(lemmatized_words)
```

5. **Apply Stemming (Optional):**

   Optionally, apply stemming to your words to reduce them to their root forms. Stemming helps to standardize different forms of a word to a common base, which can be useful for text visualization. However, whether to use stemming or lemmatization depends on your specific needs and preferences for the visualization.

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()

stemmed_words = []

for w in filtered_sentence:
    rootWord = ps.stem(w)
    stemmed_words.append(rootWord)
print(stemmed_words)
```

6. **Find the Frequency Distribution:**

   Determine the frequency distribution of your list of words using `nltk.FreqDist()`. This function computes how often each word appears in your text. For instance, if you have a list of stemmed words, you can pass this list to `nltk.FreqDist(stemmed_words)` to create a frequency distribution object. This object will store the count of occurrences for each unique word, enabling you to analyze which terms are most common in your text. By examining this frequency distribution, you can gain insights into the prevalent themes and keywords in your dataset.

```python
frequency_dist = nltk.FreqDist(stemmed_words)
frequency_dist
```

7. **Filter Words by Length:**

   To refine your list further, keep only those words that are longer than three characters. Short words or symbols like "n't," "'re," "!" and "?" often do not contribute significantly to the analysis and can be excluded. This can be done by applying a condition to your frequency distribution, where you create a dictionary `large_words` that includes only those words from `frequency_dist` with lengths greater than three characters. The resulting dictionary, `large_words`, will contain words with their corresponding frequencies, omitting the shorter, less meaningful terms. This step helps to focus on more substantial and relevant words for visualization and analysis.

```python
large_words=dict([k,v]for k,v in frequency_dist.items() if len(k)>3)
large_words
```

##**Task 2: Visualising the distribution of words using NLTK and plot()**


To visualize the frequency distribution of words in your text, you can use the `plot()` function provided by NLTK. After filtering and processing your words, you first create a frequency distribution object using `nltk.FreqDist(large_words)`, where `large_words` is a dictionary containing words longer than three characters and their frequencies. By calling `frequency_dist.plot(30, cumulative=False)`, you generate a plot that displays the top 30 most frequent words in your dataset. The `cumulative=False` argument ensures that the plot shows the raw frequencies rather than cumulative counts. This visual representation helps in quickly identifying the most common words and understanding their distribution within the text, making it easier to analyze and interpret the data.

```python
frequency_dist = nltk.FreqDist(large_words)
frequency_dist
frequency_dist.plot(30,cumulative=False)
```

## **Task 3: Visualizing Word Distribution with Word Cloud and Matplotlib:**

To create a visually intuitive representation of word frequency distribution, you can generate a word cloud using the `WordCloud` class from the `wordcloud` library. This technique provides a graphical depiction where the size of each word reflects its frequency, making it easy to identify prominent terms at a glance.

First, import the necessary libraries: `WordCloud` from the `wordcloud` package and `matplotlib.pyplot` for plotting. You then create a `WordCloud` object with specified parameters, such as `max_font_size` to set the maximum size of the words, `max_words` to limit the number of words displayed, and `background_color` to set the background color of the word cloud.

Using the `generate_from_frequencies()` method, you pass in the `frequency_dist` object, which contains the word frequencies. This method generates the word cloud based on these frequencies.

Finally, you use `matplotlib` to display the word cloud. `plt.figure()` initializes a new figure, and `plt.imshow(wordcloud, interpolation="bilinear")` renders the word cloud image in the plot. The `interpolation="bilinear"` argument ensures smooth rendering of the image. This visualization provides a clear and engaging way to see which words are most prevalent in your text data.

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

wordcloud = WordCloud(max_font_size=50, max_words=100,background_color="black").generate_from_frequencies(frequency_dist)

plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
```

You can also explore various customizations for generating word clouds, including different shapes, colors, and font sizes. For more details on how to adjust these parameters, refer to the WordCloud documentation available at: [WordCloud Documentation](https://amueller.github.io/word_cloud/generated/wordcloud.WordCloud.html).
